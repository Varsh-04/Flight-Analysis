```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_excel(r"C:\Users\varsh\Downloads\Flight_details.2.xlsx")
df
```

```python
df.info()
df.describe()
```

```python
flight_counts = df.groupby(['Year', 'Month']).size().reset_index(name='Number_of_Flights')
flight_counts
```

```python

plt.figure(figsize=(12, 6))
for year in flight_counts['Year'].unique():
    subset = flight_counts[flight_counts['Year'] == year]
    plt.plot(subset['Month'], subset['Number_of_Flights'], marker='o', label=year)

plt.title('Number of Flights by Month and Year')
plt.xlabel('Month')
plt.ylabel('Number of Flights')
plt.xticks(range(1, 13)) 
plt.legend(title='Year')
plt.grid()
plt.show()
```

```python
route_counts = df.groupby(['Source', 'Destination']).size().reset_index(name='Flight_Count')
route_counts
```

```python
popular_routes = route_counts.sort_values(by='Flight_Count', ascending=False)
popular_routes
```

```python
top_routes = popular_routes.head(10)
plt.figure(figsize=(12, 6))
plt.barh(top_routes['Source'] + ' to ' + top_routes['Destination'], top_routes['Flight_Count'], color='skyblue')
plt.xlabel('Number of Flights')
plt.title('Top 10 Popular Routes')
plt.gca().invert_yaxis()
plt.show()
```

```python
df['Actual_Duration'] = (df['Arrival_hours'] - df['Dep_hours']) / 60  
print(df[['Duration_hours', 'Duration_min', 'Actual_Duration']].head())
```

```python
df['Planned_Duration'] = df['Duration_hours'] * 60 + df['Duration_min']
df['Planned_Duration']
```

```python
df['Estimated_Delay'] = df['Actual_Duration'] - df['Planned_Duration']
df['Estimated_Delay']
```

```python
def delay_category(delay):
    if delay <= 0:
        return 'On-time'
    elif delay <= 15:
        return 'Moderate Delay'
    else:
        return 'Severe Delay'

df['Delay_Category'] = df['Estimated_Delay'].apply(delay_category)

print(df[['Source', 'Destination', 'Planned_Duration', 'Actual_Duration', 'Estimated_Delay', 'Delay_Category']].head())
```

```python
route_delays = df.groupby(['Source', 'Destination'])['Estimated_Delay'].mean().reset_index()
route_delays = route_delays.sort_values(by='Estimated_Delay', ascending=False)

print(route_delays.head(10))
```

```python
hourly_delays = df.groupby('Dep_hours')['Estimated_Delay'].mean()

plt.figure(figsize=(10, 5))
plt.plot(hourly_delays.index, hourly_delays.values, marker='o')
plt.xlabel('Hour of Day')
plt.ylabel('Average Delay (minutes)')
plt.title('Average Delay by Time of Day')
plt.grid()
plt.show()
```
